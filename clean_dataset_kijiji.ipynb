<h1><b>Cleaning Dataset</b></h1>

```python
#impoprt libraries
import numpy as np
import pandas as pd

#import dataset
kijiji_df = pd.read_csv("data_rav4_kijiji.csv")
```

```python
#view the first lines
kijiji_df.head()
```

```python
#viewing the dataset's info
kijiji_df.info()
```

```python
#describe
kijiji_df.describe(include='all')
```

```python
#cheking for duplicates
kijiji_df[kijiji_df.duplicated()]
```

```python
#84 rows duplicates, let's keep only the first
kijiji_df = kijiji_df.drop_duplicates(keep='first')
```

```python
kijiji_df.shape
```

```python
#cheking 'transmission'
kijiji_df['transmission'].value_counts()
```

```python
#after analysis, I decided to drop 2 columns: 'transmission' and 'trim'
#'transmission': because only one ad is "manual", so it does not impact the data
#'trim': because there is not a pattern, eache seller can write anything
kijiji_df = kijiji_df.drop(columns=['transmission', 'trim'])
```

```python
#cheking in the 'title' ads that is not selling, but looking for a car
kijiji_df[kijiji_df['title'].str.contains("wanted", case=False)==True]
```

```python
#remove these ads
kijiji_df = kijiji_df[kijiji_df['title'].str.contains("wanted", case=False)==False]
```

```python
#cheking in the 'title' ads that is not mention 'RAV4'
kijiji_df[kijiji_df['title'].str.contains("rav4|rav 4|rav\-4|Race 4|Rave 4|Rev4|rv 4", case=False)==False]
```

```python
#removing these rows
kijiji_df = kijiji_df[kijiji_df['title'].str.contains("rav4|rav 4|rav\-4|Race 4|Rave 4|Rev4|rv 4", case=False)==True]
```

```python
#cheking missing values
kijiji_df.isna().sum()
```

```python
#ads without price does not work for us
kijiji_df = kijiji_df.dropna(subset=['price'])
```

```python
#cheking missing values
kijiji_df.isna().sum()
```

```python
#cheking null values in kilometers
kijiji_df[kijiji_df['kilometers'].isna()]
```

```python
#Because these cars are new, so kilometers = 0
#lets cheking another new cars
kijiji_df[kijiji_df['condition']=='New']
```

```python
#only two cars with condition new.
#in this case, let drop this two ads and drop the column 'condition', because our focus is in used car
kijiji_df = kijiji_df.dropna(subset=['kilometers'])
```

```python
kijiji_df['condition'].value_counts()
```

```python
kijiji_df = kijiji_df.drop(columns=['condition'])
```

```python
#cheking missing values
kijiji_df.isna().sum()
```

```python
#cheking values in colour
kijiji_df['colour'].value_counts()
```

```python
kijiji_df[kijiji_df['colour'].isna()]
```

```python
#The best choice is use 'other'
kijiji_df.loc[kijiji_df['colour'].isna(), 'colour'] = "Other"
```

```python
#cheking missing values in Fuel
kijiji_df[kijiji_df['fuel'].isna()]
```

```python
#we can see the ads 1036 is a hybrid car, so change this one to Hybrid and the other to Gasoline
kijiji_df.loc[1036, 'fuel'] = "Hybrid-Electric"
```

```python
kijiji_df.loc[kijiji_df['fuel'].isna(), 'fuel'] = "Gasoline"
```

```python
kijiji_df.isna().sum()
```

```python
#cheking missing values in 'drivettrain"
kijiji_df[kijiji_df['drivetrain'].isna()]
```

```python
#we can see the ads 492 is a 4x4, so change this one to Hybrid and the other to "All-wheel drive (AWD)"
kijiji_df.loc[492, 'drivetrain'] = "4 x 4"
```

```python
kijiji_df.loc[kijiji_df['drivetrain'].isna(), 'drivetrain'] = "All-wheel drive (AWD)"
```

```python
kijiji_df.isna().sum()
```

```python
#No more missing values, lets checking data types
kijiji_df.info()
```

```python
#clean the column 'kilometers'
#remove ',' and change the type from object to float
kijiji_df['kilometers'] = kijiji_df['kilometers'].str.replace(',', '')
kijiji_df['kilometers'] = kijiji_df['kilometers'].astype(float)
```

```python
#change the type of 'date_posted
kijiji_df['date_posted'] = pd.to_datetime(kijiji_df['date_posted'], format='%Y-%m-%d')
```

```python
#looking for outliers at price
price_count = kijiji_df['price'].value_counts()
price_count.sort_index().head(10)
```

```python
price_count.sort_index(ascending=False).head(10)
```

```python
kijiji_df[kijiji_df['price'] < 3000]
```

```python
kijiji_df.loc[kijiji_df['price'] == 1, 'price'] = np.nan
kijiji_df = kijiji_df.dropna(subset=['price'])
```

```python
#cheking postal code
kijiji_df['postal_code'].value_counts()
```

```python
#some postal code are not correct, lets fix it
#removing spaces
kijiji_df['postal_code'] = kijiji_df['postal_code'].str.replace(' ', '')
```

```python
#checking postal code less than 6 chacteres 
kijiji_df[kijiji_df['postal_code'].str.len() < 6]
```
