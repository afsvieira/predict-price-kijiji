Collecting data from kijiji ads about the following car:<br>

<b>Maker:</b> Toyota<br>
<b>Model:</b> RAV4<br>
<b>Location:</b> Ontario<br>

```python
#import necessary libraries
import requests
import time
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#base URL for kijiji website
base_url = "https://www.kijiji.ca"

#url for the first page search (Toyota rav4 in Ontario)
page_url = base_url + "/b-cars-trucks/ontario/toyota-rav4/c174l9004a54a1000054"
```

```python
#get links from ads list
ads_links = []
response = requests.get(page_url)
soup = BeautifulSoup(response.text, 'lxml')
ads = soup.find_all("div", attrs={"class": ["search-item showcase top-feature", "search-item showcase regular-ad", "search-item top-feature", "search-item regular-ad"]})
for ad in ads:
    ads_links.append(base_url + ad["data-vip-url"])
    
while soup.find_all("a", attrs={'title': 'Next'}):
    next = soup.find_all("a", attrs={"title": "Next"})
    for page in next:
        page_url = base_url + page['href']
    response = requests.get(page_url)
    soup = BeautifulSoup(response.text, 'lxml')
    ads = soup.find_all("div", attrs={"class": ["search-item showcase top-feature", "search-item showcase regular-ad", "search-item top-feature", "search-item regular-ad"]})
    for ad in ads:
        ads_links.append(base_url + ad["data-vip-url"])
    print("Links from " + page_url + " added successfully!")

print(len(ads_links))
    


```

```python
#create a dataframe to store our results
df = pd.DataFrame(columns=["title", "price", "condition", "year", "trim", "colour", "transmission", "drivetrain", "fuel", "kilometers", "postal_code", "date_posted", "url"])
```

```python
#create a dataframe to store our results

df = pd.DataFrame(columns=["title", "price", "condition", "year", "trim", "colour", "transmission", "drivetrain", "fuel", "kilometers", "postal_code", "date_posted", "url"])

#lopping to feed the dataframe
for i, ad in enumerate(ads_links):
    
    time.sleep(1.5)
    response = requests.get(ad)
    soup = BeautifulSoup(response.text, 'lxml')
    
    #get title
    try:
        title = soup.find("h1").text
    except AttributeError:
        title = ""
    
    #get price
    try:
        price = soup.find("span", attrs={'itemprop': 'price'})['content']
    except AttributeError:
        price = ""
    except TypeError:        
        price = ""
        pass
    
    #get condition
    try:
        condition = soup.find("dd", attrs={'itemprop': 'itemCondition'}).text
    except AttributeError:
        condition = ""
    
    #get year
    try:
        year = soup.find("dd", attrs={'itemprop': 'vehicleModelDate'}).text
    except AttributeError:
        year = ""
    
    #get trim
    try:
        trim = soup.find("dd", attrs={'itemprop': 'vehicleConfiguration'}).text
    except AttributeError:
        trim = ""
        
    #get colour
    try:
        colour = soup.find("dd", attrs={'itemprop': 'color'}).text
    except AttributeError:
        colour = ""
    
    #get transmission
    try:
        transmission = soup.find("dd", attrs={'itemprop': 'vehicleTransmission'}).text
    except AttributeError:
        transmission = ""
    
    #get drivetrain
    try:
        drivetrain = soup.find("dd", attrs={'itemprop': 'driveWheelConfiguration'}).text
    except AttributeError:
        drivetrain = ""
    
    #get fuel
    try:
        fuel = soup.find("dd", attrs={'itemprop': 'fuelType'}).text
    except AttributeError:
        fuel = ""
    
    #get kilometers
    try:
        km = soup.find("dd", attrs={'itemprop': 'mileageFromOdometer'}).text
    except AttributeError:
        km = ""
        
    #get postal code
    try:
        postal_code = soup.find("span", attrs={'itemprop': 'address'}).text[-7:]
    except AttributeError:
        postal_code = ""
    
    #get date posted
    try:
        date = soup.find("div", attrs={'itemprop': 'datePosted'})['content']
    except AttributeError:
        date = ""
    except TypeError:
        date = ""
        pass
    
    df = df.append({
        "title": title,
        "price": price,
        "condition": condition,
        "year": year,
        "trim": trim,
        "colour": colour,
        "transmission": transmission,
        "drivetrain": drivetrain,
        "fuel": fuel,
        "kilometers": km,
        "postal_code": postal_code,
        "date_posted": date,
        "url": ad},
        ignore_index=True
    )
    
    print("Ad number " + str(i) + " added successfuly in Dataframe.")
    
```
