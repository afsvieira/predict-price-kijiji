```python
#import librires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#import data to DataFrame
kijiji_df = pd.read_csv("data_rav4_kijiji_cleaned.csv")
```

```python
kijiji_df.head(10)
```

```python
kijiji_df.dtypes
```

```python
kijiji_df.describe(include='all')
```

```python
kijiji_df = kijiji_df.drop(columns=['Unnamed: 0'])
```

```python
kijiji_df['year'] = pd.to_datetime(kijiji_df['year'], format='%Y')
```

```python
kijiji_df['year'] = pd.DatetimeIndex(kijiji_df['year']).year
```

```python
kijiji_df.describe(include='all')
```

```python
#ads by year
cars_per_year = kijiji_df['year'].value_counts().sort_index()
```

```python
cars_per_year.plot.bar(figsize=(20,8), color="red")
plt.grid(linestyle='--', axis='y')
plt.title("Cars being sold by year of manufacture")
plt.xlabel("Year")
plt.ylabel("Number of cars")
```

```python
#ads per fuel
cars_per_fuel = kijiji_df['fuel'].value_counts()
cars_per_fuel.plot.pie(figsize=(20,10), autopct='%1.2f%%', startangle=180)
plt.title('Fuel Type')
plt.legend()
```

```python
#year x price
import plotly.express as px
fig = px.box(kijiji_df, x='year', y='price', title='Year x Price')
fig.show()
```

```python
#KM x Price x Drivetrain
px.scatter(kijiji_df, x='kilometers', y='price', color='drivetrain', title='Kilometers x Price x Drivetrain')
```

```python
#cars x colour
cars_per_year = kijiji_df['colour'].value_counts()
cars_per_year.plot.bar(figsize=(20,8), color="red")
plt.grid(linestyle='--', axis='y')
plt.title("Cars being sold by colour")
plt.xlabel("Colour")
plt.ylabel("Number of cars")
```

```python
kijiji_df['date_posted'] = pd.to_datetime(kijiji_df['date_posted'], format='%Y-%m-%d')
kijiji_df['days_on_market'] = (pd.to_datetime('today', format='%Y-%m-%d') - kijiji_df['date_posted']).dt.days
```

```python
groupby_colour_mean = kijiji_df.groupby(by='colour').mean()
```

```python
px.line(groupby_colour_mean, y='price', title='Average Price By colour')
```

```python
groupby_year_mean = kijiji_df.groupby(by='year').mean()
```

```python
groupby_year_mean
```

```python
px.scatter(groupby_year_mean, y='price', size='kilometers', title="Average price by Year and Kilomentes(AVG)")
```

```python
min_price_year = kijiji_df.groupby(by='year').min()
```

```python
px.bar(min_price_year, y='price', color='kilometers', title="Lower Price by year")
```

```python
max_price_year = kijiji_df.groupby(by='year').max()
```

```python
px.bar(max_price_year, y='price', color='kilometers', title="Higher Price by year")
```
